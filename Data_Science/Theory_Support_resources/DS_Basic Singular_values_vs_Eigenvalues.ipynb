# [Singular Values vs. Eigenvalue](https://www.youtube.com/watch?v=Ed6CSJbyVak)

## SVD:  
M = U (left singular vectors) * Σ (diagonal matrix, singular values) * V^T (right singular vectors).  
<br>
Note: U and V are orthonormal (U * U^T = Identity matrix) and Σ * Σ^T == Σ^2 (since Σ is diagonal matrix) 


**NOTE: M*M^T and M^T*M  are all square matrix, has Eigenvalue/vectors**
