## [Singular Value Decomposition](https://www.youtube.com/watch?v=HAJey9-Q8js&ab_channel=ritvikmath)
**KEY: Eigen Decomposition only apply to SQUARE Matrix**  
SVD generalize that
**Note: "Rank" is the number of independent columns**

Express the **Matrix M** as sum of several **rank-1 independent vectors**

So if a **Rank P matrix** can be decomposed to P **Rank 1 matrix** then add up 

It has its value when has to represent a Large Matrix


Suppose we had a matrix M with 100 rows and 10 columns. Normally requires 1000 space to store these data.  
But suppose the rank of this matrix is 2, sigma 1 and sigma 2. We can simply represent this matrix with less spaces.
