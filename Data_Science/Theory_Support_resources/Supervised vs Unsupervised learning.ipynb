# Supervised machine learning Vs Unsupervised machine learning

## [Intro to ML resources(Google)](https://developers.google.com/machine-learning/problem-framing/cases)
## supervised machine learning  
A case when we have both features(the matrix X) and the labels(the vector y). 
Common categories: Classification, regression

## unsupervised machine learning 
A case we don't have the laebls(the vector y)  
Common categories: Clustering, Association rule learning
