# Dimension Reduction
----
**What is Dimension Reduction? Why do we care about it?**  
<br>
When dataset is multidimensional, is hard to interpret and hard to visualize it. Also additional dimensions spread the data even further making high dimensional data extremely sparse. We care about it, because it is difficult to use machine learning in sparse spaces.

## [Principal Component Analysis (PCA)](https://www.youtube.com/watch?v=FgakZw6K1QQ&ab_channel=StatQuestwithJoshStarmer)
---
**Whta is PCA?**
<br>


PCA convert the **correlations among the samples** into a 2-D graph.  
<br>
It is a special type of SVD where the singular values correspond to the eigenvectors and the values of the diagonal matrix are the squares of the eigenvalues. We use these features as they are statistically descriptive.  
<br>
**Noet: clustering based on minizing the linear distances is the same maximizing the linear correlations**

----
**What it means about its axes?**
<br>
The PC1 and PC2 means, among all the PC(Principal Components), find the two PCs that accounts the highest variances. The two axes (i.e., the x-axis and y-axis) imply the variation in data along these two PCs.  

NOTE: SS = sum for squared distances from projected points to origin
