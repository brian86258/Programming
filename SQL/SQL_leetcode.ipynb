# 626. Exchange Seats


---
[Sol](https://leetcode.com/problems/exchange-seats/discuss/104702/A-solution-without-using-union-and-order-by):
KEY: usage of **when .... then...**, 類似 **if .... : ... ***
<br>
select id,  
case   
    when id %2 = 0 then (select student from seat where id = (i.id-1))  
    when id %2 != 0 and id < (select count(*) from seat) then (select student from seat where id = (i.id+1)) 
    else student  
end as student  
from seat i

where e1.Salary > e2.Salary

# 184. Department Highest Salary

KEY, Using **group by** and **having** will return **only ONE ROW**. Hence, cannot be used in this scenario. 
**[No subquery sol](https://leetcode.com/problems/department-highest-salary/discuss/53610/My-best-solution-super-clean-no-subquery-no-Max)**
    
SOL:  
select d.Name as Department, e.Name as Employee, e.salary as Salary  
from employee e join department d on e.DepartmentId = d.Id  
where e.salary = (select max(salary) from Employee e2 where e2.DepartmentId = d.Id)

# 178. Rank Scores
[Rank Introduction](https://www.cnblogs.com/52xf/p/4209211.html)
  

SOL:  
select score, Dense_Rank() OVER(order by score DESC) as `Rank`  
from Scores   
order by  score DESC
